## **Lagrange Interpolation**

### **Polynomial Equations**

```python
from sympy import symbols,simplify,Poly
import numpy as np
from sympy.plotting import plot
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
def Lagrange(x_list, y_list):
  x = symbols('x')
  y = 0
  for k in range(len(x_list)):
    mult = 1
    for j in range(len(x_list)):
      if j != k:
          mult = mult* ((x-x_list[j])/(x_list[k]-x_list[j]))
    y += mult*y_list[k]
  return y
```

```python
## 3rd degree polynomial:
x_list3 = [0, 1.03, 1.76, 2.29]
y_list3 = [2.718, 0.368, 1.712, 2.3]

poly3 = simplify(Lagrange(x_list3,y_list3))
print("P(X) = {}".format(poly3))
plot(poly3)
```

```python
## 4th degree polynomial:
x_list4 = [0, 0.6, 1.03, 1.76, 2.29]
y_list4 = [2.718, 0.797, 0.368, 1.712, 2.3]

poly4 = simplify(Lagrange(x_list4,y_list4))
print("P(X) = {}".format(poly4))
plot(poly4)
```

```python
## 5th degree polynomial:
x_list5 = [0, 0.6, 1.03, 1.39, 1.76, 2.29]
y_list5 = [2.718, 0.797, 0.368, 0.597, 1.712, 2.3]

poly5 = simplify(Lagrange(x_list5,y_list5))
print("P(X) = {}".format(poly5))
plot(poly5)
```

```python
## 6th degree polynomial:
x_list6 = [0, 0.6, 1.03, 1.39, 1.76, 2.09, 2.29]
y_list6 = [2.718, 0.797, 0.368, 0.597, 1.712, 2.718, 2.3]

poly6 = simplify(Lagrange(x_list6,y_list6))
print("P(X) = {}".format(poly6))
plot(poly6)
```

### **Plotting f(x) and interpolated polynomials**

```python
def Interpolate_Polynomial(X, x_list, y_list):
  estimated_y = []
  for x in X:
    poly = 0
    for k in range(len(x_list)):
      mult = 1
      for j in range(len(x_list)):
        if j != k:
            mult = mult* ((x-x_list[j])/(x_list[k]-x_list[j]))
      poly += mult*y_list[k]
    estimated_y.append(poly)
  return estimated_y
```

```python
x = np.linspace(0,2.3,100)
# 3rd degree polynomial:
interpolation3 = Interpolate_Polynomial(x, x_list3, y_list3)

# 4th degree polynomial:
interpolation4 = Interpolate_Polynomial(x, x_list4, y_list4)

# 5th degree polynomial:
interpolation5 = Interpolate_Polynomial(x, x_list5, y_list5)

# 6th degree polynomial:
interpolation6 = Interpolate_Polynomial(x, x_list6, y_list6)

plt.plot(x, np.exp(np.cos(3*x)))
plt.plot(x, interpolation3)
plt.plot(x, interpolation4)
plt.plot(x, interpolation5)
plt.plot(x, interpolation6)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['f(x)','3rd degree polynomial','4th degree polynomial','5th degree polynomial','6th degree polynomial'])
plt.title('f(x) and interpolated polynomials')
plt.show()
```

### **Error Graphs**

```python
fx = np.exp(np.cos(3*x))
error3 = abs(interpolation3 - fx)
error4 = abs(interpolation4 - fx)
error5 = abs(interpolation5 - fx)
error6 = abs(interpolation6 - fx)
plt.plot(x, error3)
plt.plot(x, error4)
plt.plot(x, error5)
plt.plot(x, error6)
plt.xlabel('x')
plt.ylabel('error')
plt.legend(['3rd degree polynomial','4th degree polynomial','5th degree polynomial','6th degree polynomial'])
plt.title('Error values  for each polynomial')
plt.show()
```
